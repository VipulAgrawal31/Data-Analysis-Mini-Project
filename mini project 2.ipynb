```python
import pandas as pd 
df = pd.read_csv('AppleStore.csv')
d=pd.DataFrame(df)
d




```

```python
k1=list(d['prime_genre'])
k2=list(d['user_rating'])

```

```python
games=0
nongames=0
j=0
count1=0
count2=0
for i in range(len(k1)):
    if k1[j]=='Games':
        count1+=1
        games=games+k2[j]
        j=j+1
    else:
        nongames+=k2[j]
        count2+=1
        j=j+1
print(games)
print(nongames)
print("================")
avg1=games/count1
avg2=nongames/count2
print("Games avg ",avg1)
print("Nongames avg " ,avg2)
```

 ### Price =0  - Free apps
 - Price <20 - Affordable apps
 - Price<50 - Expensive apps
 - Price >50 - Very expensive

```python
 Price=list(d["price"])
j=0
count1=0
count2=0
count3=0
count4=0
for i in range(len(Price)):
    if Price[j]==0:
        count1+=1
        j+=1
    elif Price[j]<20:
        count2+=1
        j+=1
    elif Price[j]<50:
        count3+=1
        j+=1
    elif Price[j]>50:
        count4+=1
        j+=1
print(count1)
print(count2)
print(count3)
print(count4)     
```

```python
import matplotlib.pyplot as plt
labels = ['Very expensive','Affordable apps','Expensive apps','Free apps']
sizes = [count4,count2,count3,count1]
colors = ['violet', 'green', 'yellow', 'brown']
explode = (0.1, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
